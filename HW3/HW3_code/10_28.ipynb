```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
N1=100
A=np.zeros([N1,N1])
for i in range(0,N1):
    A[i,i]=6
for i in range(0,N1-1):
    A[i,i+1]=-4
    A[i+1,i]=-4
for i in range(0,N1-2):
    A[i,i+2]=1
    A[i+2,i]=1
```

```python
condiA=np.linalg.norm(A,2)*np.linalg.norm(np.linalg.inv(A),2)
#condiA=np.linalg.cond(A)
print(condiA)
print('A is ill-conditioned')
```

```python
N2=1000
perturbrange=1e-08
memory_Aperturb=np.zeros([N2,1])
E1,V1=np.linalg.eig(A)
E1=np.sort(E1)

for i in range(0,N2):
    Aperturb=perturbrange*np.random.rand(N1,N1)
    E2,V2=np.linalg.eig(A+Aperturb)
    E2=np.sort(E2)
    memory_Aperturb[i,0]=np.linalg.norm(E1-E2)

plt.plot(memory_Aperturb,'ro')
plt.title('Eigenvalue perturbation of A')
plt.show()

print('eigenvalues are not ill-conditioned')
```
