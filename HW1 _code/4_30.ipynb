##### First, import python libraries useful in this calculation

```python
import numpy as np
from sympy import Matrix,pprint
```

#### Second, define elementary row operation function and determinant function based on elementary row operaton.  
Elementary row operation in this function does not include normalization of 1st non-zero entries of each row.

```python
#This function applies for (m,n) matrix. Elementary row op, with initial entries not normalized to 1
def elemrowop(A,I,m,n):
    tempA=A
    tempI=I
    ref=0
    for i in range(0,n):
        if(ref>=m):
            break
        if ((tempA[ref:m,i]==np.zeros((1,m-ref))).all()):
            continue
        for j in range(ref,m):
            if(tempA[j,i]==0):
                continue
            else:
                break
        E1=np.eye(m,k=0,dtype=float)
        if(ref!=j):
            E1[ref,ref]=0;E1[j,j]=0;E1[ref,j]=1;E1[j,ref]=1
        tempA=E1@tempA
        tempI=E1@tempI
        E2=np.eye(m,k=0,dtype=float)
        for j in range(ref+1,m):
            E2[j,ref]=-((tempA[j,i])/(tempA[ref,i]))
        tempA=E2@tempA
        tempI=E2@tempI
        ref+=1
    return tempA,tempI

#This function only applies to square matrix
def elemrowop2(A,I,n):
    tempA=A
    tempI=I
    if((np.diag(A)==np.zeros((1,n))).any()):
        print('A is singular matrix')
        return 0,0
    else:
        for i in range(0,n):
            E=np.eye(n,k=0,dtype=float)
            for j in range(0,n-i-1):
                E[j,n-1-i]=-tempA[j,n-1-i]/tempA[n-1-i,n-1-i]
            tempA=E@tempA
            tempI=E@tempI
    return tempA,tempI

#This function only applies to square matrix
def diagop(A,I,n):
    tempA=A
    tempI=I
    diag=np.eye(n,k=0,dtype=float)
    for i in range(0,n):
        if(tempA[i,i]==0):
            diag[i,i]=0
        else:
            diag[i,i]=1/tempA[i,i]
    tempA=diag@tempA
    tempI=diag@tempI    
    return tempA,tempI

#Function for determinant of a matrix
def determinant(A,n):
    tempA=A
    ref=0
    sgn=0
    det=1
    for i in range(0,n):
        if(ref>=n):
            break
        if ((tempA[ref:n,i]==np.zeros((1,n-ref))).all()):
            continue
        for j in range(ref,n):
            if(tempA[j,i]==0):
                continue
            else:
                break
        E1=np.eye(n,k=0,dtype=float)
        if(ref!=j):
            E1[ref,ref]=0;E1[j,j]=0;E1[ref,j]=1;E1[j,ref]=1
            sgn+=1
        tempA=E1@tempA
        E2=np.eye(n,k=0,dtype=float)
        for j in range(ref+1,n):
            E2[j,ref]=-((tempA[j,i])/(tempA[ref,i]))
        tempA=E2@tempA
        ref+=1
    
    #diagonal product
    for i in range(0,n):
        det*=tempA[i,i]
    det=det*pow(-1,sgn)
    
    return det

def hilbert(n):
    A=np.zeros((n,n))
    for i in range(0,n):
        for j in range(0,n):
            A[i,j]=1/(i+j+1)
    return A
```

### Third, calculate determinant of Hilbert matrix of n=5,10,15,25 using functions defined above

### Below code is answer for (a)

```python
H5=hilbert(5)
H10=hilbert(10)
H15=hilbert(15)
H25=hilbert(25)

#detH5=determinant(H5,5)
#detH10=determinant(H10,10)
#detH15=determinant(H15,15)
#detH25=determinant(H25,25)

detH5=Matrix(H5).det()
detH10=Matrix(H10).det()
detH15=Matrix(H15).det()
detH25=Matrix(H25).det()

print('determinant of hilbert matrix of n=5 is %E'%(detH5))
print('determinant of hilbert matrix of n=10 is %E'%(detH10))
print('determinant of hilbert matrix of n=15 is %E'%(detH15))
print('determinant of hilbert matrix of n=25 is %E'%(detH25))

print('\nAs n gets bigger, elements in the inverse matrix tends to get smaller toward 0')
```
