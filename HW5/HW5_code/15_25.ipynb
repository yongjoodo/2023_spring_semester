```python
import numpy as np
import scipy as sp
```

```python
n=8
A=np.zeros((n,n))
for i in range(0,n):
    for j in range(0,n):
        A[i,j]=1/(i+j+1)
```

```python
rkA=np.linalg.matrix_rank(A)
print('rank of Hilbert matrix H8 is',rkA)
```

```python
U,S,Vt=np.linalg.svd(A)
print('U=\n',U);print('S=\n',np.diag(S));print('V=\n',Vt.T)
```

```python
diagS=np.zeros([1,8])
for i in range(0,8):
    diagS[0,i]=S[i]
print(diagS)
print('Singular value gets really small...')
```

```python
condA=S[0]/S[7]
print('Condition number of H8 is',condA)
print('2-norm of the inverse of hilb(8) has very large elements,and its 2-norm is large')
```
