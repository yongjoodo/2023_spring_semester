```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.array([[1.3, -12.3],
               [1.8, -8.5 ],
               [2.3, -5.6 ],
               [3.6, -2.3 ],
               [4.9, -0.5 ],
               [5.3,  1.3 ],
               [6.5,  1.5 ]])
```

```python
def Vandermonde(x,n):
    V=np.zeros([len(x),n+1])
    for i in range(0,n+1):
        V[:,i]=x**i
    return V
```

```python
V=Vandermonde(data[:,0],1)
linear=np.linalg.solve(V.transpose()@V,V.transpose()@data[:,1])
print(linear)
```

```python
V=Vandermonde(data[:,0],2)
quadratic=np.linalg.solve(V.transpose()@V,V.transpose()@data[:,1])
print(quadratic)
```

```python
x=np.array(range(1,8))
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('LSM, linear regression, and quadratic regression')
plt.scatter(data[:,0],data[:,1],c='black',marker='x')
plt.plot(x,linear[0]+linear[1]*x)
plt.plot(x,quadratic[0]+quadratic[1]*x+quadratic[2]*x**2)

plt.show()
```
