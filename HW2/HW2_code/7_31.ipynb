```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

A_1=np.array([[0.6, 1, 6, -1, 5],
              [0, 0.6, 1, 1, 0],
              [0, 0, 0.6, 1, 3],
              [0, 0, 0, 0.6, 1],
              [0, 0, 0, 0, -0.7]])

index_1=np.zeros([1,51])
norm_1=np.zeros([1,51])
tempA_1=np.eye(5)
for i in range (0,51):
    index_1[0,i]=i
    norm_1[0,i]=np.linalg.norm(tempA_1,2)
    tempA_1=tempA_1@A_1

plt.plot(index_1,norm_1,'bx')
plt.title('First case of A')
plt.show()
```
